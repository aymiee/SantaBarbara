<h3> 1.1 Background

The location of a restaurant will determine its success. In a wrong location, you will not attract the amount of customers needed to stay in business. <br>

<h3> 1.2 Problem

Which neighborhood in Santa Barbara, CA would be the optimal location for a Roman-Style pizzeria - the first of its kind as there are other pizzerias in town.  

The objective of this capstone project is to locate the optimal neighborhood for a new Roman-Style pizzeria operation using Foursquare API <br>
and census data. I will look at spending power, rent, distribution of ethnic group, and competition, across each neighbourhood.
